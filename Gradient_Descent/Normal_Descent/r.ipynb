Gradient Descent

```python
import numpy as np

class GradientDescent:

    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.b = 0
    
    def coefficients(self):
        return self.weights
    
    def intercept(self):
        return self.b
    
    
    def fit(self, X_train, y_train):

        # Number of rows and columns
        rows, cols = X_train.shape

        # Initialize one weight for each feature
        self.weights = np.zeros(cols)
        self.b = 0

        # Gradient Descent
        for i in range(self.iterations):

            # Prediction
            y_hat = X_train @ self.weights + self.b

            # Calculate gradients
            dw = (-2 / rows) * np.sum(X_train.T @ (y_train - y_hat))
            db = (-2 / rows) * np.sum(y_train - y_hat)

            # Update weights and bias
            self.weights = self.weights - self.learning_rate * dw
            self.b = self.b - self.learning_rate * db

            # Print progress
            if i % 100 == 0:
                print(f"Iteration {i}")
                print(f"Weights = {self.weights}")
                print(f"Bias = {self.b}")
                print("-" * 40)

    def predict(self, X_test):
        return X_test @ self.weights + self.b
```

```python
# Train custom model
my_model = GradientDescent(
    learning_rate=0.01,
    iterations=1000
)

my_model.fit(X, y)

print("\n========== Gradient Descent ==========")
print("Intercept :", my_model.intercept())
print("Coefficient :", my_model.coefficients())

# Prediction
y_pred = my_model.predict(X)
# accuracy of my models is
r2 = r2_score(y, y_pred)

print("R-squared:", r2)
```
