```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

class NormalDescent:
    def __init__(self, learning_rate=0.01, iterations=100):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.m = 0
        self.b = 0

    def fit(self, X, y, plot_every=10):
        X = np.array(X)
        y = np.array(y)
        n = len(X)
        
        # Store history to see convergence
        self.m_history = []
        self.b_history = []
        self.loss_history = []
        
        # Setup plot (only one figure)
        plt.figure(figsize=(12, 5))
        
        for i in range(self.iterations):
            # Current predictions
            y_hat = self.m * X + self.b
            
            # Calculate loss (MSE)
            loss = np.mean((y - y_hat) ** 2)
            
            # Gradients
            dm = (-2 / n) * np.sum(X * (y - y_hat))
            db = (-2 / n) * np.sum(y - y_hat)
            
            # Update parameters
            self.m -= self.learning_rate * dm
            self.b -= self.learning_rate * db
            
            # Store history
            self.m_history.append(self.m)
            self.b_history.append(self.b)
            self.loss_history.append(loss)
            
            # Plot every few iterations
            if i % plot_every == 0 or i == self.iterations - 1:
                plt.clf()
                
                # Subplot 1: Data and current fit
                plt.subplot(1, 2, 1)
                plt.scatter(X, y, color='blue', alpha=0.6, label='Data Points')
                x_line = np.array([X.min(), X.max()])
                y_line = self.predict(x_line)
                plt.plot(x_line, y_line, 'r-', linewidth=2, 
                        label=f'Iteration {i}: m={self.m:.3f}, b={self.b:.3f}')
                plt.xlabel('CGPA')
                plt.ylabel('Salary Offered (LPA)')
                plt.title('Gradient Descent Learning')
                plt.legend()
                
                # Subplot 2: Loss over time
                plt.subplot(1, 2, 2)
                plt.plot(self.loss_history, 'g-', alpha=0.7)
                plt.axvline(x=i, color='r', linestyle='--', alpha=0.3)
                plt.xlabel('Iteration')
                plt.ylabel('Mean Squared Error')
                plt.title(f'Loss at iteration {i}: {loss:.2f}')
                plt.grid(True, alpha=0.3)
                
                plt.tight_layout()
                plt.pause(0.01)  # Small pause for update
        
        plt.show()
        print(f"\n{'='*50}")
        print(f"FINAL RESULTS after {self.iterations} iterations:")
        print(f"  Slope (m):     {self.m:.6f}")
        print(f"  Intercept (b): {self.b:.6f}")
        print(f"  Final Loss:    {loss:.4f}")
        print(f"{'='*50}")

    def predict(self, X):
        X = np.array(X)
        return self.m * X + self.b
```

```python
normal_model = NormalDescent(
    learning_rate=0.001,
    iterations=10
)

normal_model.fit(X, y)

# Final prediction and R² score
from sklearn.metrics import r2_score

y_pred_normal = normal_model.predict(X)

print("Gradient Descent R² Score:",
      r2_score(y, y_pred_normal))

print("Sklearn R² Score:",
      r2_score(y, predict_y))
```
