```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()



# Load data
df = pd.read_csv("/Users/mohddanish/Desktop/Machine_Learning/MachineLearning/Gradient_Descent/Normal_Descent/Batch_gradient_descent/placement.csv")

# Display first few rows
print("Dataset Preview:")
print(df.head())
print("\nDataset Info:")
print(df.info())
```

```python
# Use only one feature
X = df.drop("Salary_Offered_LPA", axis=1).values
x_standard = scaler.fit_transform(X)
X = x_standard
y = df["Salary_Offered_LPA"].values
```

```python
# Train sklearn model
lr_model = LinearRegression()
lr_model.fit(X, y)


print("========== Scikit-Learn ==========")
print("Intercept :", lr_model.intercept_)
print("Coefficient :", lr_model.coef_)

y_pred = lr_model.predict(X)

r2 = r2_score(y, y_pred)

print("R-squared:", r2)
```

```python
import numpy as np

class StochasticGradientDescent:

    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.b = 0

    def coefficients(self):
        return self.weights

    def intercept(self):
        return self.b

    def fit(self, X_train, y_train):

        rows, cols = X_train.shape

        # Initialize weights
        self.weights = np.zeros(cols)
        self.b = 0

        # Number of epochs
        for i in range(self.iterations):

            # Loop through every training example
            for j in range(rows):

                # Pick one random sample
                random_index = np.random.randint(0, rows)

                # Don't overwrite the dataset
                x_i = X_train[random_index]     # Shape = (cols,)
                y_i = y_train[random_index]     # Scalar

                # Prediction for one sample
                y_hat = x_i @ self.weights + self.b

                # Compute gradients (NO division by rows)
                dw = -2 * x_i * (y_i - y_hat)
                db = -2 * (y_i - y_hat)

                # Update parameters immediately
                self.weights = self.weights - self.learning_rate * dw
                self.b = self.b - self.learning_rate * db

            # Print progress every 100 epochs
            if i % 100 == 0:
                print(f"Epoch {i}")
                print("Weights:", self.weights)
                print("Bias:", self.b)
                print("-" * 40)

        return self

    def predict(self, X_test):
        return X_test @ self.weights + self.b
```

```python
stochastic_models = StochasticGradientDescent()
stochastic_models.fit(X, y)

print("========== Stochastic Gradient Descent ==========")
print("Intercept :", stochastic_models.intercept())
print("Coefficient :", stochastic_models.coefficients())

y_pred = stochastic_models.predict(X)

r2 = r2_score(y, y_pred)

print("R-squared:", r2)
```
