```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Categorical Data Univarite Analysis


```python
# how many peoples are travelling
total_people = df.shape[0]
print(f'Total people travelling: {total_people}')
```

```python
df['Survived'].value_counts()
```

```python
# count plot for the survived column
sns.countplot(x='Survived', data=df)
plt.title('Count of Survived vs Not Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# countplot for the Pclass column
sns.countplot(x='Pclass', data=df)
plt.title('Count of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# using pie chart for the Pclass column
# pclass_counts = df['Pclass'].value_counts()
# plt.figure(figsize=(6, 6))
# plt.pie(pclass_counts, labels=pclass_counts.index, autopct='%1.1f%%')
# plt.title('Distribution of Passenger Classes')
# plt.axis('equal')
# plt.show()

# using pie chart for the Survived column
survived_counts = df['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Survival')
plt.axis('equal')
plt.show()

```

### Numerical Data Univarite Analysis


```python
# numerical data univarite analysis
# histogram for the Age column
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.show()
```
