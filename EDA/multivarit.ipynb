```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_iris = pd.read_csv('iris.csv')
df_titanic = pd.read_csv('train.csv')
df_tips = pd.read_csv('tip.csv')
```

```python
df_iris.head()
```

```python
df_tips.head()
```

ScatterPlot (Numerical-Numerical
)

```python
sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue=df_tips['sex'])
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

Bar Plot (Numerical - Categorical)


```python
# bar plot for numerical and categorical data 
# age and plcass 
# and make sure y - axis always have the average of the numerical data and x - axis have the categorical data
sns.barplot(x='Pclass', y='Age', data=df_titanic, hue=df_titanic['Sex'])
plt.title('Average Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Age')
plt.show()
```

BoxPlot (Numerical-Categorical)


```python
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Pclass', y='Age', data=df_titanic, hue=df_titanic['Sex'])
    plt.title('Boxplot of Age by Passenger Class')
    plt.xlabel('Passenger Class')        
    plt.show()
```

Distplot (Numerical - Categorical)

```python
sns.kdeplot(df_titanic[df_titanic['Survived'] == 1]['Age'], label='Survived')
sns.kdeplot(df_titanic[df_titanic['Survived'] == 0]['Age'], label='Died')
```
