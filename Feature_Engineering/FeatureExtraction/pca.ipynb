```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
X = np.array([
    [1, 2, 3],
    [2, 1, 0],
    [4, 5, 6]
])

print("Original Dataset:\n")
print(X)
print(f"size of the dataset: {X.shape}")
```

```python
mean = np.mean(X, axis=0)

print("\nMean of each column:\n")
print(mean)
```

```python
X_centered = X - mean

print("\nMean Centered Data:\n")
print(X_centered)
print(f"shape of the dataset: {X_centered.shape}")
```

```python
cov_variance_matrix = np.cov(X_centered)
print("\nCovariance Matrix:\n")
print(cov_variance_matrix)
print(f"shape of the dataset: {cov_variance_matrix.shape}")
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_variance_matrix)
print("\nEigenvalues:\n")
print(eigenvalues)
print("\nEigenvectors:\n")
print(eigenvectors)

sort_eigenvalues = np.argsort(eigenvalues)[::-1]
# argsort returns the indices that would sort the array. [::-1] reverses the order to get descending order.
print("\nEigenvectors in descending order:\n")
print(sort_eigenvalues)
sorted_eigenvectors = eigenvectors[:, sort_eigenvalues]
print(sorted_eigenvectors)


```
