```python
X_train, X_test, y_train, y_test = train_test_split(df[['Age', 'EstimatedSalary']], df['Purchased'], test_size=0.2, random_state=42)
```

```python
Scaler = StandardScaler()
X_train_scaled = Scaler.fit_transform(X_train)
X_test_scaled = Scaler.transform(X_test)
```

```python
df.head()
```

```python
# merge the standardized columns into the original dataframe
df['StandardizedAge'] = Scaler.fit_transform(df[['Age']])
df['StandardizedEstimatedSalary'] = Scaler.fit_transform(df[['EstimatedSalary']])
df.head()
```

```python
# scatter plot of original features
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=df['Age'], y=df['EstimatedSalary'], hue=df['Purchased'])
plt.title('Original Features')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')

# scatter plot of standardized features
plt.subplot(1, 2, 2)
sns.scatterplot(x=df['StandardizedAge'], y=df['StandardizedEstimatedSalary'], hue=df['Purchased'])
plt.title('Standardized Features')
plt.xlabel('Standardized Age')
plt.ylabel('Standardized Estimated Salary')
plt.tight_layout()
plt.show()
```

Why Feature Scaling is important


```python
lr = LogisticRegression()
lr.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test_scaled)
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy of Logistic Regression with Standardized Features: {accuracy:.2f}')
```

```python
# Without feature scaling, the accuracy might be lower due to the model being sensitive to the scale of the features. By standardizing the features, we ensure that they are on a similar scale, which can improve the performance of many machine learning algorithms, including logistic regression.
lr_unscaled = LogisticRegression()
lr_unscaled.fit(X_train, y_train)
```

```python
y_pred_unscaled = lr_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

print(f'Accuracy of Logistic Regression without Standardization: {accuracy_unscaled:.2f}')
```
